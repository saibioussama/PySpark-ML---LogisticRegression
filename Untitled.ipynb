```python
from pyspark.sql import SparkSession
import pandas as pd
import matplotlib.pyplot as plt

spark = SparkSession.builder.appName('ml-bank').getOrCreate()
original_data = spark.read.csv('ibm-hr-analytics-attrition.csv', header = True, inferSchema = True)

columnNames = ["Age","Attrition" ,"Department","Gender" ,"JobSatisfaction" ,"MonthlyIncome","TotalWorkingYears","OverTime","PerformanceRating"]
df = original_data.select(columnNames)
pandas = df.toPandas()
 
```

```python
dummies = pd.get_dummies(pandas["Gender"])
pandas = pd.concat([dummies,pandas],axis='columns')

dummies = pd.get_dummies(pandas["Department"])
pandas = pd.concat([dummies,pandas],axis='columns')

dummies = pd.get_dummies(pandas["Attrition"])
pandas = pd.concat([dummies,pandas],axis='columns')

dummies = pd.get_dummies(pandas["OverTime"])
pandas = pd.concat([dummies,pandas],axis='columns')

pandas = pandas.drop(["Gender","Department","Attrition","OverTime"],axis='columns')

pandas
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = pandas.drop(["PerformanceRating"],axis='columns')
X
```

```python
y = pandas.PerformanceRating
y
```

```python
model.fit(X,y)
```

```python
model.predict([[1,0,1,0,0,1,0,0,1,34,3,4404,6]])
```

```python
model.score(X,y)
```
